# Ex1 - Filtering and Sorting Data

Step 1. Import the necessary libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Step 2. Import the dataset from this address.
(https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv)

Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('/content/chipotle.tsv.txt',sep='\t')
chipo
```

Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: x[1:])
```

```python
chipo['item_price'] = chipo['item_price'].astype('float')
```

```python
len(chipo[chipo['item_price'] > 10]['item_name'])
```

Step 5. What is the price of each item?

```python
chipo['price_per_item'] = chipo['item_price']/chipo['quantity']

chipo.head()
```

```python
chipo[['item_name','price_per_item']]
```

Step 6. Sort by the name of the item




```python
chipo.sort_values(by='item_name')
```

Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by='item_price',ascending=False).head(1).loc[:,['quantity']]
```

```python
chipo.sort_values(by='item_price',ascending=False).iloc[:1,1]
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
len(chipo[chipo['item_name']=='Veggie Salad Bowl'])
```

Step 9. How many times did someone order more than one Canned Soda?


```python
len(chipo[(chipo['item_name']=='Canned Soda')& (chipo['quantity']>1)])
```
